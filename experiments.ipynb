```python
from tqdm.auto import tqdm 

with open("experiments/log", 'w') as w:
    pass

affiliation_combinations = ["khalifa university", "ku", "kustar", "kust", "masdar", "petroleum institute"]

scholar_ids = []

for name in tqdm(df["name"]):
    #try to find by name from faculty directory
    first_search = list(scholarly.search_author(name))
    n_first_search = len(first_search)
    #if no scholar pages found or multiple pages found
    if n_first_search == 1:
        scholar_ids.append(first_search[0]["scholar_id"])
        #print("first", "https://scholar.google.com/citations?hl=en&user=" + first_search[0]["scholar_id"])
        #TODO: build confirmation
    else:
        #if more than one profile are found, try to narrow down by looking at affilations
        if n_first_search > 1:
            #try to look for name + combination
            for profile in first_search:
                found = False
                for combination in affiliation_combinations:
                    #print(combination, profile["affiliation"])
                    if combination in profile["affiliation"].lower():
                        scholar_ids.append(profile["scholar_id"])
                        #print("second", "https://scholar.google.com/citations?hl=en&user=" + profile["scholar_id"])
                        found = True
                    else:
                        pass
                        #print(name, profile)
                    if found:
                        break
                if found:
                    break
    if not found:
        print("!=1")
        with open("experiments/log", 'a') as w:
            w.writelines(name + "----" + str(first_search) + "\n")
    print("----------------------------")

                    
                 
```
